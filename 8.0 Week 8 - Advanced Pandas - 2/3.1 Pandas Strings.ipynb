##### Split function

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df['Fare'] = df['Fare'].astype("string")
df['dollar_value'] = df['Fare'].str.split(".").str.get(0)
df['cent_value'] = df['Fare'].str.split(".").str.get(1)
```

```python
df.head(2)
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)

print(df['title'].value_counts())
df.head()
```

##### Replace function

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

##### Filtering

```python
# filtering
# startswith/endswith
df[df['first_name'].str.endswith('A')]
# isdigit/isalpha...
df[df['first_name'].str.isdigit()]
```

#### Regular Expression

```python
df_main = pd.DataFrame()
vowel = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]

for i in df.index:

    df_sub = df[df.index==i]
    first_element = df_sub['lastname'].str.get(0)[i]
    last_element = df_sub['lastname'].str.get(-1)[i]
    
    if (first_element in vowel) | (last_element in vowel):
        df_sub['vowel_check'] = "True"
    else:
        df_sub['vowel_check'] = "False"
        
    df_main = pd.concat([df_main, df_sub])
```

```python
df_main[df_main['vowel_check'] == 'True']
```

```python
# applying regex
# contains
# search john -> both case
df[df['first_name'].str.contains('john',case=False)]

# find lastnames with start and end char vowel
df[~df['lastname'].str.contains('^[aeiouAEIOU].*[aeiouAEIOU]$')]
```
