```python
# Converting a integer to string
number = int(input('enter the number'))

digits = '0123456789'
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10

print(result)
```

Time Complexity is <b>O(logn)</b>

```python
# Finding sum and product
L = [1,2,3,4]

sum = 0
for i in L:
  sum = sum + i

product = 1
for i in L:
  product = product*i

print(sum,product)
```

Time Complexity is <b>O(n)</b>

```python
# Printing all combination of items in two list
A = [1,2,3,4]
B = [5,6,7,8]
for i in A:
  for j in B:
    print(i,j)
```

```python
# Reversing a list
L = [1,2,3,4,5]

for i in range(0,len(L)//2):
  other = len(L) - i -1
  temp = L[i]
  L[i] = L[other]
  L[other] = temp

print(L)
```

Time Complexity is <b>O(n)</b>

```python
# Nested loops
n = 10
k = 0;
for i in range(n//2,n):
  for j in range(2,n,pow(2,j)):
        k = k + n / 2;

print(k)
```

```python
# Sum of each digit of a number
n = 345

sum = 0
while n>0:
  sum = sum + n%10
  n = n // 10

print(sum)
```

Time Complexity is <b>O(logn)</b>

```python
# fibonacii series
def fib(n):
  if n == 1 or n == 0:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```
